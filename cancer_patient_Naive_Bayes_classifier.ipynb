```python
import time
start_time = time.time()
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sklearn

fileName = "/content/cancer patient data sets.csv"

print("SetUp Complete")
```

```python
df = pd.read_csv(fileName, encoding='latin-1')
df
```

```python
# Index Column now refers to patient

print('\n')
df.drop("Patient Id", axis=1, inplace=True)

# Cleaning Column Names
df.rename(columns=str.lower, inplace=True)
df.rename(columns={col: col.replace(" ", "_") for col in df.columns}, inplace=True)

# Display Data After Cleaning
display(df)
print('\n')
```

```python
#Chek for Null values
df.isnull().sum()
# there is no null values
```

```python
# We have to change the type of Level column from string to number to claculate the naive byase

df["level"].replace({'High': 2, 'Medium': 1, 'Low': 0}, inplace=True) # replace High to 2, Medium to 1 and Low to 0
print('Cancer Levels: ', df['level'].unique())
```

```python
# Setting the target
# Setting Target

X = df.drop(columns='level') # Fetures
Y = df.level # traget

print('\n')
display(X, Y)
print('\n')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size= 0.2, random_state=40)
# 20% testing and 80% training
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

model = GaussianNB()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
import numpy as np

def calculate_accuracy():
    # Ensure the input data are NumPy arrays
    true_labels = np.array(y_test)
    predicted_labels = np.array(y_pred)

    # Calculate the error rate
    error_rate = np.mean(true_labels != predicted_labels)

    # Calculate the accuracy from the error rate
    accuracy = 1 - error_rate

    return accuracy * 100  # return accuracy as a percentage



accuracy_percentage = calculate_accuracy()
print(f'Accuracy: {accuracy_percentage:.2f}%')
```

```python
end_time = time.time()  # Store the current time again

runtime = end_time - start_time  # Calculate the difference
print(f'Runtime: {runtime} seconds')
```
